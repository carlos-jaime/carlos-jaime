```python
from google.cloud import bigquery
import json
import os
import io

from deepdiff import DeepDiff
```

```python
GCP_CREDENTIALS = os.environ["GOOGLE_APPLICATION_CREDENTIALS"]

BQ_CLIENT = bigquery.Client.from_service_account_json(GCP_CREDENTIALS)
```

#### raw_table

```python
# Table Id
raw_table =  {"project_id":"peya-data-origins-pro",
                 "dataset_id":"raw_dmarts",
                 "table_id":"customer_orders"
        }



raw_table_id = f"{raw_table['project_id']}.{raw_table['dataset_id']}.{raw_table['table_id']}"
```

```python
# Obtiene la metadata de la tabla raw_table
table = BQ_CLIENT.get_table(raw_table_id)

# Obtiene el schema de la tabla y lo transforma en un diccionario
f = io.StringIO("", newline='')

BQ_CLIENT.schema_to_json(table.schema, f)

raw_schema = json.loads(f.getvalue())

new_schema = []

for items in raw_schema:
    new_items = {}
    for key in items:
        if key == 'fields':
            print('fields')
        elif key != 'description':
            new_items[key] = items[key]
            #print(f'{key} : {items[key]}')
    new_schema.append(new_items)
            
    
#raw_schema.pop('description')

#raw_schema
new_schema
```

```python
def remove_key_from_schema(schema,field_to_remove,new_schema=[]):

    for items in schema:
        new_items = {}
        for key in items:
            if key == 'fields':
                remove_key_from_schema(items[key],field_to_remove,new_schema)
            elif key != field_to_remove:
                new_items[key] = items[key]
                
        new_schema.append(new_items)
    
```

```python
test_new_schema = []
remove_key_from_schema(schema=raw_schema, field_to_remove='description', new_schema=test_new_schema)

new_schema
```
